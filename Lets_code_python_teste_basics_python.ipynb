```python
CONFIRMADOS = 0
OBITOS = 1
RECUPERADOS = 2
ATIVOS = 3
DATA = 4
```

```python
def get_datasets(y, labels):
  if type(y[0]) == list:
      datasets = []
      for i in range (len(y)):
          datasets.append({
              'label': labels[i],
              'data' : y[i]
          })
      return datasets
  else:
    return [
            {
                'label':lables[0],
                'data': y
            }
    ]
```

```python
def set_title(title=''):
  if title != '':
    display = 'true'
  else:
    display = 'false'
    return {
      'title': title,
      'display': display
  }
```

```python
def creat_chart(x, y, lables, kind ='bar',  title =''):
  datasets = get_datasets(y, labels)
  options = set_title(title)
  chart = {
      'type': kind,
      'data': {
          'labels':x,
          'datasets': datasets
      },
      'options': options
  }
  return chart
```

```python
def get_api_chart(chart):
  url_base = 'https://quickchart.io/chart'
  resp = r.get(f'{url_base}?c={str(chart)}')
  return resp.content
```

```python
def save_image(path, content):
  with open(path, 'wb') as image:
    image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
  img_pil = Image.open(path)
  display(img_pil)
```

```python
y_data_1 = []
for obs in final_data[1::10]:
  y_data_1.append(obs[CONFIRMADOS])

y_data_2 = []
for obs in final_data[1::10]:
  y_data_1.append(obs[RECUPERADOS])

labels = ['Confirmados','Recuperados']

x=[]
for obs in final_data[1::10]:
  x.append(obs[DATA].strftime('%d/%m/%Y'))

chart = creat_chart(x,[y_data_1, y_data_2], labels, title ='Grafico Confirmados x Recuperados')
chart_content = get_api_chart(chart)
save_image('meu-primeiro-grafico.png', chart_content)
display_image('meu-primeiro-grafico.png')
```

```python
from urllib.parse import quote
```

```python
def get_api_qrcode(link):
  text = quote(link) #parsing do link para url
  url_base = 'https://quickchart.io/qr'
  resp = r.get(f'{url_base}?text={text}')
  return resp.content
```
